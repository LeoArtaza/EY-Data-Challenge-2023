```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import numpy as np
import seaborn as sns
import os
```

```python
subs = []
for path, subdirs, files in os.walk("."):
    for name in files:
        if 'csv' in name and 'Version' in name and 'check' not in name and 'ada' not in name and 'mix' not in name:
            print(name)
            subs.append(pd.read_csv(os.path.join(path, name))['Predicted Rice Yield (kg/ha)'])
```

```python
preds = pd.concat(subs, axis=1).round(-2).astype(int)
preds.columns = [0.16, 0.25, 0.51, 0.45, 0.52, 0.49, 0.53, 0.57, 0.49, 0.46, 0.55, 0.55, 0.48]
preds = preds.sort_index(axis=1, ascending=False).iloc[:, :-2]
preds
```
